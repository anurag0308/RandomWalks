```python
import networkx as nx
import matplotlib.pyplot as plt
import numpy as np
import random
#for (u, v) in G.edges():
        #G.edges[u,v]['weight'] = random.random()
```

```python
def randomgraph(nodes,probability):
    G = nx.gnp_random_graph(nodes,probability, directed = True)
    nx.draw(G,with_labels=True)
    return G 
```

```python
node = 5
prob = 0.5
```

```python
G=randomgraph(node,prob)
len(G)
```

```python
r = 1
G = randomgraph(node,prob)
c = len(G.nodes)
b = np.ones( (r,c))
InitialProbMatrix = b/c
InitialProbMatrix


```

```python
AdjecencyMatrix = nx.to_numpy_matrix(G)
AdjecencyMatrix



```

```python
row_sum = AdjecencyMatrix.sum(axis=1)
row_sum
```

```python
TransitionProbMatrix = AdjecencyMatrix/row_sum
TransitionProbMatrix
```

```python
def PageRank(noi,InitialProbMatrix,TransitionProbMatrix,threshold):
    for i in range(noi):
        InitialProb = InitialProbMatrix
        PageRank= np.matmul(InitialProb,TransitionProbMatrix)
        NextProb = PageRank
        terminating_condition = np.sum(np.absolute(NextProb-InitialProb))
        if (terminating_condition < threshold):
            InitialProb = PageRank
    return (PageRank,i)
```

```python
k,i=PageRank(1000,InitialProbMatrix,TransitionProbMatrix,0.05)
```
